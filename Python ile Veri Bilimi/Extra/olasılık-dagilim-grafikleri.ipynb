## Büyük Sayılar Kanunu

```python
import numpy as np
import matplotlib.pyplot as plt

# Zar atma fonksiyonu
def roll_dice():
    return np.random.randint(1, 7)

# Yüz sayıları
n = 1000

# Örneklem sayıları
num_samples = [100, 1000, 10000]

fig, axs = plt.subplots(nrows=1, ncols=3, figsize=(15,5), facecolor='whitesmoke')

for i, ns in enumerate(num_samples):

    # Ortalama değerleri saklamak için dizi oluştur
    sample_means = np.zeros(n)

    # n kez zar at, her defasında ortalamayı kaydet
    for j in range(n):
        sample = np.array([roll_dice() for k in range(ns)])
        sample_means[j] = np.mean(sample)

    # Grafik çiz
    axs[i].hist(sample_means, bins=30, color='#967bb6', alpha=0.8)
    axs[i].set_xlabel('Ortalama')
    axs[i].set_ylabel('Frekans')
    axs[i].set_title(f'{ns} Örnekleme için Ortalama Dağılımı')

plt.show()
```

## Kolmogorov-Smirnov Testi

```python
import numpy as np
from scipy.stats import kstest, norm

# Normal Dağılımdan Örnek Veri Seti Oluşturma
mu = 0
sigma = 1
n = 100
data = np.random.normal(mu, sigma, n)

# Kolmogorov-Smirnov Testi
D, p = kstest(data, 'norm', args=(mu, sigma))

# Test Sonucu
print(f"Test İstatistiği (D): {D}")
print(f"p-değeri (p): {p}")
if p > 0.05:
    print("Veri seti Normal dağılıma uyuyor (Ho kabul edildi)")
else:
    print("Veri seti Normal dağılıma uymuyor (Ho reddedildi)")
```

## Mann-Whitney U Testi

```python
import numpy as np
from scipy.stats import mannwhitneyu

# İki Örneklem Veri Seti Oluşturma
data1 = np.random.normal(0, 1, 50)
data2 = np.random.normal(1, 1, 50)

# Mann-Whitney U Testi
U, p = mannwhitneyu(data1, data2)

# Test Sonucu
print(f"Test İstatistiği (U): {U}")
print(f"p-değeri (p): {p}")
if p > 0.05:
    print("İki örneklem arasında anlamlı bir fark yoktur (Ho kabul edildi)")
else:
    print("İki örneklem arasında anlamlı bir fark vardır (Ho reddedildi)")
```

## ANOVA Testi (Tek Yönlü)

```python
import numpy as np
from scipy.stats import f_oneway

# Üç Örneklem Veri Seti Oluşturma
data1 = np.random.normal(0, 1, 50)
data2 = np.random.normal(1, 1, 50)
data3 = np.random.normal(2, 1, 50)

# ANOVA Testi
f, p = f_oneway(data1, data2, data3)

# Test Sonucu
print(f"Test İstatistiği (F): {f}")
print(f"p-değeri (p): {p}")
if p > 0.05:
    print("Gruplar arasında anlamlı bir fark yoktur (Ho kabul edildi)")
else:
    print("Gruplar arasında anlamlı bir fark vardır (Ho reddedildi)")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pareto

# Pareto dağılım parametreleri
b = 1 # shape parametresi
k = 3 # scale parametresi

# X değerleri aralığı
x = np.linspace(0.1, 10, 100)

rv = pareto(b, scale=k)

pdf = rv.pdf(x)

fig, ax = plt.subplots(facecolor='whitesmoke')
ax.plot(x, pdf, color='#967bb6', alpha=0.8)
ax.set_xlabel('X')
ax.set_ylabel('f(x)')
ax.set_title('Pareto Dağılımı (b=1, k=3)')

plt.text(0.8, 0.9, r'$f(x) = \frac{b k^b}{x^{b+1}}$',
         fontsize=12, transform=ax.transAxes, bbox=dict(facecolor='#967bb6', alpha=0.5))

plt.show()
```

## Cauchy Dağılımı

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import cauchy

# Cauchy dağılım parametreleri
x0 = 0 # merkez noktası
gamma = 1 # yarı genişlik

x = np.linspace(-10, 10, 100)

rv = cauchy(loc=x0, scale=gamma)

pdf = rv.pdf(x)

fig, ax = plt.subplots(facecolor='whitesmoke')
ax.plot(x, pdf, color='#967bb6', alpha=0.8)
ax.set_xlabel('X')
ax.set_ylabel('f(x)')
ax.set_title('Cauchy Dağılımı (x0=0, γ=1)')

plt.text(0.7, 0.9, r'$f(x) = \frac{1}{\pi \gamma [1 + (\frac{x - x_0}{\gamma})^2]}$',
         fontsize=12, transform=ax.transAxes, bbox=dict(facecolor='#967bb6', alpha=0.5))

plt.show()
```

## Rayleigh Dağılımı

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import rayleigh

# Rayleigh dağılım parametreleri
sigma = 1 # ölçek parametresi

x = np.linspace(0, 10, 100)

rv = rayleigh(scale=sigma)

pdf = rv.pdf(x)

fig, ax = plt.subplots(facecolor='whitesmoke')
ax.plot(x, pdf, color='#967bb6', alpha=0.8)
ax.set_xlabel('X')
ax.set_ylabel('f(x)')
ax.set_title('Rayleigh Dağılımı (σ=1)')

plt.text(0.7, 0.9, r'$f(x) = \frac{x}{\sigma^2} e^{-\frac{x^2}{2\sigma^2}}$',
         fontsize=12, transform=ax.transAxes, bbox=dict(facecolor='#967bb6', alpha=0.5))

plt.show()
```

## Laplace Dağılımı

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import laplace

# Laplace dağılım parametreleri
mu = 0 # ortanca
b = 1 # ölçek parametresi

x = np.linspace(-10, 10, 100)

rv = laplace(loc=mu, scale=b)

pdf = rv.pdf(x)

fig, ax = plt.subplots(facecolor='whitesmoke')
ax.plot(x, pdf, color='#967bb6', alpha=0.8)
ax.set_xlabel('X')
ax.set_ylabel('f(x)')
ax.set_title('Laplace Dağılımı (μ=0, b=1)')

plt.text(0.74, 0.9, r'$f(x) = \frac{1}{2b} e^{-\frac{|x-\mu|}{b}}$',
         fontsize=12, transform=ax.transAxes, bbox=dict(facecolor='#967bb6', alpha=0.5))

plt.show()
```

## Gumbel Dağılımı

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gumbel_r

# Gumbel dağılım parametreleri
mu = 0  # ortalama
beta = 1  # ölçek parametresi

x = np.linspace(-10, 10, 100)

rv = gumbel_r(loc=mu, scale=beta)

pdf = rv.pdf(x)

fig, ax = plt.subplots(facecolor='whitesmoke')
ax.plot(x, pdf, color='#967bb6', alpha=0.8)
ax.set_xlabel('X')
ax.set_ylabel('f(x)')
ax.set_title('Gumbel Dağılımı (μ=0, β=1)')

plt.text(0.67, 0.9, r'$f(x) = \frac{1}{\beta} e^{-(\frac{x-\mu}{\beta}+e^{-\frac{x-\mu}{\beta}})}$',
         fontsize=12, transform=ax.transAxes, bbox=dict(facecolor='#967bb6', alpha=0.5))

plt.show()
```
